```python
# assess tagset lengths
orig_tag_dir = '/home/ubuntu/praxi/tag_improvements/orig_tags'
# 
new_tag_dir = '/home/ubuntu/praxi/tag_improvements/edited_tags'

from os import listdir
from os.path import isfile, join
orig_fnames = [f for f in listdir(orig_tag_dir) if isfile(join(orig_tag_dir, f))]
new_fnames = [f for f in listdir(new_tag_dir) if isfile(join(new_tag_dir, f))]

print(orig_fnames[0])
```

```python
# load all tagsets into a list of dictionaries
import yaml
from tqdm import tqdm

old_dics = []
new_dics = []

for name in tqdm(orig_fnames):
    full_name = orig_tag_dir + '/' + name
    with open(full_name, 'r') as stream:
        data = yaml.safe_load(stream)
    old_dics.append(data)
    
for name in tqdm(new_fnames):
    full_name = new_tag_dir + '/' + name
    with open(full_name, 'r') as stream:
        data = yaml.safe_load(stream)
    new_dics.append(data)
    
```

```python
# assess lengths
print(len(old_dics))
print(old_dics[0])
#print(len(orig_fnames))
#print(len(new_fnames))
```

```python
# count labels
old_labels = []
new_labels = []
for cur_dic in tqdm(old_dics):
    cur_label = cur_dic['label']
    if cur_label not in old_labels:
        old_labels.append(cur_label)
        
for cur_dic in new_dics:
    cur_label = cur_dic['labels']
    if cur_label not in new_labels:
        new_labels.append(cur_label)
        
print(len(old_labels))
print(len(new_labels))
```

```python
#print(new_dics[0])
# list of 0's
old_tag_sizes = [0] * 83
new_tag_sizes = [0] * 83

for i in old_dics:
    # get index from label name
    label_name = i['label']
    idx = old_labels.index(label_name)
    # length of the dictionary
    dic_length = len(i['tags'])
    if new_tag_sizes[idx] == 0:
        old_tag_sizes[idx] = dic_length
    
for i in new_dics:
    # get index from label name
    label_name = i['labels']
    idx = new_labels.index(label_name)
    # length of the dictionary
    dic_length = len(i['tags'])
    if new_tag_sizes[idx] == 0:
        new_tag_sizes[idx] = dic_length
    
print(old_tag_sizes)
print(new_tag_sizes)
```

```python
# see which ones are the shortest:
for lab, length in zip(old_labels, old_tag_sizes):
    if length < 10 or lab == 'wget':
        print(lab, length)

print("BREAK")
for lab, length in zip(new_labels, new_tag_sizes):
    if length < 10 or lab == 'wget':
        print(lab, length)
        
```
