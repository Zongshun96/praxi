```python
from itertools import combinations
from statistics import mean

# Jaccard Index
def jac(s1,s2):
    """the jaccard index between 2 sets"""
    s_union = s1.union(s2)
    s_inter = s1.intersection(s2)

    len_union = len(s_union)
    if not len_union:
        return 0

    return len(s_inter)*1.0/len_union

def jac_full(dict0, dict1):
    """Comparison taking into account keys and items"""
    l0 = dict0['tags']
    l1 = dict1['tags']
    
    d0 = {}
    d1 = {}

    for item in l0:
        labfreq = item.split(':')
        d0[labfreq[0]] = float(labfreq[1])
    
    for item in l1:
        labfreq = item.split(':')
        d1[labfreq[0]] = float(labfreq[1])
        
    set_keys0 = set(d0.keys())
    set_keys1 = set(d1.keys())

    key_commonality = jac(set_keys0, set_keys1)

    set0 = set(d0.items())
    set1 = set(d1.items())

    item_commonality = jac(set0, set1)

    #ok, so now we give a proximity on key match, even if key + data dont match
    return key_commonality  #*.9 + 0.1 * item_commonality


def assess_sim(dic_list):
    # compare all dictionaries in the list to each other, return
    # average jaccard index
    combos = list(combinations(dic_list, 2))
    jac_scores = [jac_full(i[0], i[1]) for i in combos]
    #print(jac_scores)
    return mean(jac_scores)

def assess_sim_2(dl1, dl2):
    # create list of all combinations of one element from list 1, one element from l2
    pairs = [] #[] for i in range(len(dl1)*len(dl2))]
    for el_out in dl1:
        for el_in in dl2:
            pair = [el_out, el_in]
            pairs.append(pair)
    #print(pairs[0])
    #print(len(pairs))
    jac_scores = [jac_full(i[0], i[1]) for i in pairs]
    return mean(jac_scores)
    
```

```python
def run_tests(names):
    # takes a list of names, separates them into versions, get tagsets, and run comparisons
    # return cross-comparison, self-comparison, and avg length
    pieces = names[0].split('.')
    a_name = pieces[0]
    print("Application name:", a_name)
    
    versions = []
    for name in names: 
        pieces = name.split('-')
        ext_len = len(pieces[-1]) + 1
        just_vers = name[:-ext_len]
        if just_vers not in versions:
            versions.append(just_vers)
    print("Versions:")
    for v in versions:
        print(v)
    # separate into groups and 
    # run the tests!
    sep_list = [ [] for i in range(len(versions))]
    
    for name in names: 
        pieces = name.split('-')
        ext_len = len(pieces[-1]) + 1
        just_vers = name[:-ext_len]  
        full_name = tagset_directory + '/' + name
        with open(full_name, 'r') as stream:
            dic = yaml.safe_load(stream)
        sep_list[versions.index(just_vers)].append(dic)
        
    print("Self comparison:")
    self_comp = []
    for l in sep_list:
        print("Label:", l[0]['label'])
        js = assess_sim(l)
        print("Jaccard score:", js)
        self_comp.append(js)
    print("Uniformity: ", self_comp)
    print(" ") 
    
    print("Cross Comparison of versions:")
    combos = list(combinations(sep_list, 2))
    comp_scores = [assess_sim_2(i[0], i[1]) for i in combos]
    print(len(comp_scores))
    print(comp_scores)
    
    print("Avg Tagset Size:")
    version_means = [] 
    for l in sep_list:
        cur_vers_tot = 0 
        for el in l:
            cur_vers_tot += len(el['tags'])
        cur_vers_mean = cur_vers_tot*1.0/len(l)
        version_means.append(cur_vers_mean)
    print(version_means)
    
    print("\n")
```
